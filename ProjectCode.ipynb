```python
#to import pandas library
import pandas as pd

#To read the movies dataset
movies = pd.read_csv("movies.csv")
movies.head()

```

```python
#To read the ratings dataset
ratings = pd.read_csv("ratings.csv")
ratings.head()
```

```python
#To merge the both datasets into a single dataset
ratings = pd.merge(movies,ratings)

#to display the records
ratings.head()
```

```python
#To create a pivot table using rating values
user_ratings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')

user_ratings.head()
```

```python
#to fill the null values in the records
user_ratings = user_ratings.dropna(thresh=10, axis=1).fillna(0)

user_ratings.head()
```

```python
#create similarity matrix using pearson corelation method
item_similarity_df = user_ratings.corr(method='pearson')

item_similarity_df.head()
```

```python
#to get similar movies
def get_similar_movies(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

```python
#to recommend movies to the users based on their interest
similar_movies=pd.DataFrame()

user = [("Avengers, The (1998)",5),("Avengers: Age of Ultron (2015)",4.9)]
for movie,rating in user:
    similar_movies=similar_movies.append(get_similar_movies(movie,rating),ignore_index=True)


similar_movies = similar_movies.sum().sort_values(ascending=False)

similar_movies[2:15]


```
